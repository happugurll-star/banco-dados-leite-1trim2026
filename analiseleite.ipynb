```python
import pandas as pd
import numpy as np

tabela_total = pd.read_csv("leite_1trim2026.csv")
display(tabela_total)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.bar(tabela_total["estado"],tabela_total["qnt_cru"])
plt.xticks(rotation=90)
plt.title("Consumo de leite cu por estado")
plt.show()
display(tabela_total.columns)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.bar(tabela_total["estado"],tabela_total["qnt_industrializado"])
plt.xticks(rotation=90)
plt.title("Consumo de leite industrializado por estado")
plt.show()
display(tabela_total.columns)
```

```python
import matplotlib.pyplot as plt

top4 = tabela_total.sort_values(by="qnt_cru", ascending=False).head(4)

plt.figure(figsize=(8,5))
plt.bar(top4["estado"], top4["qnt_cru"])
plt.title("Top 4 consumidores de leite cru")
plt.xlabel("Estado")
plt.ylabel("Quantidade de leite cru")
plt.show()
```

```python
import matplotlib.pyplot as plt

top4 = tabela_total.sort_values(by="qnt_industrializado", ascending=False).head(4)

plt.figure(figsize=(8,8))
plt.pie(
    top4["qnt_cru"],
        labels = top4["estado"],
        autopct = "1%.1f%%"
        )

plt.title("Estados que mais consomem leite industrializado")
plt.show()
```
